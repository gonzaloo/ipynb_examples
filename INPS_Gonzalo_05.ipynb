## Problem 1 - Simple pendulum

```python
# Remember to give these commands first
%pylab inline
from __future__ import division
import pandas as pd
pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier
figsize(15, 5)
```

```python
# Read the data into Python
link_name = 'http://physics.info/curve-fitting/simple-pendulum.txt'
pendulum =  pd.read_csv(link_name, '\t', header = [0,1])

# Display 4 first rows
pendulum[:5]

# Get the columns from the data set
Length = pendulum['Length'].astype(int32)
Period = pendulum['Period'].astype(float)

# Plot the Period versus Length
figure()
plot(Length, Period, 'o')
title('Period (s) versus Length (cm)')

```

```python
Lenght = pendulum.iloc[:,0].astype(int32)
Period = pendulum.iloc[:,1].astype(float)
```

```python
# period of a simple pendulum  T = 2*pi*(l/g)^1/2
# 2*pi*(1/g)^1/2 is constant, so the period of a pendulum is directly proportional to the square root of its lenght.
P_t = 2*pi*sqrt(1/981)*sqrt(Length)
plot(Length, P_t, 'dr-')
plot(Length, Period, 'o')
```

```python
#The period of the pendulum is directly proportional to the lenght of it.  The relation of period-length is a square root, thus,
#it's not a linear graph.
```

## Problem 2 - Vostok

```python
# Read the data into Python
link_name = 'http://physics.info/curve-fitting/vostok.txt'
vostok =  pd.read_csv(link_name, '\t', header = [0,1])

# Display 4 first rows
vostok[:5]

# Get the columns from the data set
age = vostok['age'].astype(int32)
delta_t = vostok['∆T'].astype(float)

# Plot the dT versus age
figure()
plot(age, delta_t, 'b-')
title('age vs $\Delta T$')
xlabel('age')
ylabel('$\Delta T$')
```

### CO2

## Hubble's Law

```python
link = 'http://physics.info/curve-fitting/hubble-law.txt'
hubb =  pd.read_csv(link, '\t', header = [0,1])
hubb[:24]
```

```python
distance = hubb['Distance'].astype(float)
rv = hubb['Radial Velocity'].astype(float)
plot(distance, rv, 'go')
title('Hubble-s Law')
xlabel('Distance')
ylabel('Radial velocity')
```

```python
rad_v = 454.16 * distance - 40.784  #linear fit
plot(distance, rv, 'go')
plot(distance, rad_v, 'r-')
title('Hubble-s Law')
xlabel('Distance')
ylabel('Radial velocity')
```

```python
#The conclusion is that the Universe is expanding, because the radial velocity
#of the galaxies increases directly proportional as the distance increases.
```
