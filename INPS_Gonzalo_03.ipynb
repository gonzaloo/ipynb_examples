# Problem 1. Basic graphs

```python
k = 5
x = linspace(0, 50, 50)
y = 0*x + k
plot(x,y,'-')
title(' constant: $ y = k $')
```

```python
y = k * x
plot(x,y,'g-')
title(' linear: $ y = kx $')
```

```python
x = linspace(0.1, 10, 50)
y = k / x
plot(x,y,'r-')
title(' hyperbolic: $ y = k/x $')
```

```python
x=linspace(0, 10, 50)
y = k * (x**2)
plot(x,y,'m-')
title(' cuadratic: $ y = kx^2 $')
```

```python
y = k * (x**(1/2))
plot(x,y,'y-')
title(' rational: $ y = k \sqrt{x}) $')
```

```python
n = 5
y = k * (x**(n))
plot(x,y,'c-')
title(' polinomic: $ y = kx^n $')
```

```python
b = 15
y = k * x + b
plot(x,y,'k-')
title(' linear: $ y = kx+b $')
```

```python
a = 2
c = -9
y = a*(x**2) + b*x + c
plot(x,y,'r-')
title(' cuadratic: $ y = ax^2 + bx + c $')
```

```python
y = a*(n**(k*x))
plot(x,y,'g-')
title(' exponential: $ y = an^{kx} $')
```

```python
y = a*(n**(-k*x))
plot(x,y,'y-')
title(' exponential: $ y = an^{-kx} $')
```

```python
y = a*(1 - (n**(-k*x))) + b
plot(x,y,'r-')
title(' exponential: $ y = a(1 - n^{-kx}) + b $')
```

```python
y = a * sin(k*x + b)
plot(x,y,'g-')
title(' trigonometric: $ y = a * sin(kx + b) $')
```
