```python
#The Newton-Raphson method needs the derivative of a function to be calculated
#when h tends to be zero, the derivative is defined as f'(x) = [f(x+h)-f(x)]/h
def derivada(f,x1):
    h = 1e-10
    return ((f(x1+h)-f(x1))/h)

#To calculate the further values of x, the formula is:
# x2 = x1 - f(x1)/f'(x1)
def new_rap(f,x):
    itera = 0
    itera_res = [x]
    while(not( -1e-7 < f(x) < 1e-7)):   #checking for 7 decimals
        x = x - f(x)/derivada(f,x)
        itera_res.append(x)
        itera = itera + 1
        
        if(itera >= 100):   #infinite loop break
            break
            
    plot(range(itera+1), itera_res, 'bo-')
    xlabel("Number of iterations")
    ylabel("Result")
    grid()
    
    print "number of iterations:", itera
    return x
```

```python
#It's needed to define each equation as a function:
def y1(x):
    return 2*cos(x)-x**2

def y2(x):
    return 3*x**3 - 4*x**2 + 2*x - 9

def y3(x):
    return exp(x/2) - 5*x

def y4(x):
    return log(x) - 1/x

def y5(x):
    return sin(x) + (2*x/pi) - 1
```

```python
#1
sol1=new_rap(y1,0.8)
print 'The solution is: x =',sol1
```

```python
#2
sol2=new_rap(y2,2)
print 'The solution is: x =',sol2
```

```python
#3
sol3=new_rap(y3,6)
print 'The solution is: x =',sol3
```

```python
#4
sol4=new_rap(y4,1.6)
print 'The solution is: x =',sol4
```

```python
#5
sol5=new_rap(y5,0.6)
print 'The solution is: x =',sol5
```
